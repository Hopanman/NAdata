```python
kmdb = pd.read_csv('./NAdata/KMDB 관련/KMDB변수.csv')
```

```python
kmdb.loc[:,'영화제목'] = kmdb['영화제목'].apply(lambda x: re.sub('[^a-zA-Z0-9가-힣]','',x))
```

```python
kmdb.loc[:, '개봉일'] = pd.to_datetime(kmdb['개봉일'])
```

```python
kmdb = kmdb.rename({'영화제목':'영화명'},axis=1)
```

```python
young = pd.read_csv('NAdata/영진위 관련/영진위 영화 개봉일람(2010~2019).csv', encoding='cp949')
```
